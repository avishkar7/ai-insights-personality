```python
import pandas as pd
import numpy as np
import plotly
import plotly.express as px
import plotly.graph_objects as go
import glob
import os
import pickle
import matplotlib.pyplot as plt
import seaborn as sns


```

```python
my_path = os.path.abspath(os.path.dirname('evaluation'))
eval_dir = os.path.join(my_path, "data/evaluation//")
```

```python
files = [f for f in glob.glob(eval_dir + "**/*.p", recursive=True)]

evals = []
for f in files:
    x = pickle.load(open( f, "rb"))
    for e in x:
        evals.append(e)
```

```python
df = pd.DataFrame(evals)
df.columns = ['entries', 'vectorizer', 'alg', 'trait', 'score']
temp = df.sort_values(by=['vectorizer', 'entries', 'score'], ascending=False)
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    print(temp)
```

```python
# Create a bar plot using Seaborn and Matplotlib
plt.figure(figsize=(10, 6))
sns.barplot(x='entries', y='score', hue='vectorizer', data=df)
plt.title('Scores by Vectorizer')
plt.xlabel('Entries')
plt.ylabel('Score')
plt.xticks(rotation=45)
```

```python
# Create a grouped bar plot using Seaborn and Matplotlib
plt.figure(figsize=(12, 6))
sns.barplot(x='entries', y='score', hue='trait', data=df)
plt.title('Scores by Trait and Entries')
plt.xlabel('Entries')
plt.ylabel('Score')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a grouped bar plot using Seaborn and Matplotlib
plt.figure(figsize=(18, 10))
sns.barplot(x='vectorizer', y='score', hue='trait', data=df)
# plt.title('Scores by Trait and Entries')
plt.xlabel('Vectorizer')
plt.ylabel('Score')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a grouped bar plot using Seaborn and Matplotlib
plt.figure(figsize=(12, 6))
sns.barplot(x='alg', y='score', hue='vectorizer', data=df)
plt.title('Scores by Trait and Entries')
plt.xlabel('Algorithm')
plt.ylabel('Score')
#plt.xticks(rotation=30)
plt.show()
```

```python
# Create a grouped bar plot using Seaborn and Matplotlib
plt.figure(figsize=(12, 6))
sns.barplot(x='alg', y='score', hue='trait', data=df)
plt.title('Scores by Trait and Entries')
plt.xlabel('Algorithm')
plt.ylabel('Score')
#plt.xticks(rotation=30)
plt.show()
```

```python
filtered_df = df[df['trait'] == 'cNEU']
filtered_df = filtered_df.sort_values(by=['score'], ascending=False)
filtered_df 

# Create a bar plot to visualize the highest scores for each trait
plt.figure(figsize=(12, 6))
sns.barplot(x='alg', y='score', hue='vectorizer', data=filtered_df)
plt.title('Highest Scores by Trait and Vectorizer')
plt.xlabel('cNEU')
plt.ylabel('Score')
plt.legend(title='Vectorizer', loc='upper right')
plt.show()
```
